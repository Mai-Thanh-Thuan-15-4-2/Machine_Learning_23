#3.1. Apply StandardScaler() function to columns that contains numerical data ('age', 'balance', 'day', 'campaign', 'pdays', 'previous')

```python


names = ['age', 'balance', 'day', 'campaign', 'pdays', 'previous']
array_data = dataset[names]


scaler = StandardScaler()
scaler.fit(array_data)


scaled_data = pd.DataFrame(scaler.transform(array_data), columns=names)


dataset[names] = scaled_data


dataset.head()
```

#3.2. Apply Encode Categorical Value (OneHotEncoder) to transfrom categorical data to numerical data ('job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome')

```python
#code
onehot = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']

encoder = OneHotEncoder()

encoded_cols = encoder.fit_transform(dataset[onehot]).toarray()
encoded_df = pd.DataFrame(encoded_cols, columns=encoder.get_feature_names_out(onehot))

data_concat = pd.concat([dataset.drop(onehot, axis=1), encoded_df], axis=1)

print(data_concat)
pd.get_dummies(data_concat,columns=['age', 'balance', 'day', 'campaign', 'pdays', 'previous'],drop_first=True)
```

#3.3. Apply **Decision tree, Random forest, kNN, NaïveBayes** to preproceed dataset in the previous steps. Then compare the obtained results using **accuracy, precision, recall, f1** measures.

```python
#code
X_train, X_test, y_train, y_test = train_test_split(data_concat.iloc[:,0:7], data_concat.iloc[:,-1], test_size=0.3, random_state=42)

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
dtp = dt.predict(X_test)

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf_predicted = rf.predict(X_test)

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knnp = knn.predict(X_test)

nb = GaussianNB()
nb.fit(X_train, y_train)
nbp = nb.predict(X_test)

dt_accuracy = accuracy_score(y_test, dtp)
dt_precision = precision_score(y_test, dtp)
dt_recall = recall_score(y_test, dtp)
dt_f1 = f1_score(y_test, dtp)

rf_accuracy = accuracy_score(y_test, rf_predicted)
rf_precision = precision_score(y_test, rf_predicted)
rf_recall = recall_score(y_test, rf_predicted)
rf_f1 = f1_score(y_test, rf_predicted)

knn_accuracy = accuracy_score(y_test, knnp)
knn_precision = precision_score(y_test, knnp)
knn_recall = recall_score(y_test, knnp)
knn_f1 = f1_score(y_test,knnp)

nb_accuracy = accuracy_score(y_test, nbp)
nb_precision = precision_score(y_test, nbp)
nb_recall = recall_score(y_test, nbp)
nb_f1 = f1_score(y_test, nbp)

# Print the results
print("Results for Decision Tree : ")
print("Accuracy -", dt_accuracy)
print("Precision -", dt_precision)
print("Recall -", dt_recall)
print("F1 Score -", dt_f1)
print()
print("Results for Random Forest : ")
print("Accuracy -", rf_accuracy)
print("Precision -", rf_precision)
print("Recall -", rf_recall)
print("F1 Score -", rf_f1)
print()
print("Results for kNN : ")
print("Accuracy -", knn_accuracy)
print("Precision -", knn_precision)
print("Recall -", knn_recall)
print("F1 Score -", knn_f1)
print()
print("Results for Naive Bayes : ")
print("Accuracy -", nb_accuracy)
print("Precision -", nb_precision)
print("Recall -", nb_recall)
print("F1 Score -", nb_f1)
```
